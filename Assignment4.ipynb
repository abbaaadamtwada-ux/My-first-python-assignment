```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
filted = [num for num in numbers if num <= 0]
print(filted)
list_of_lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for sublist in list_of_lists for num in sublist]
print(flattened)
result = [tuple(a**b for b in range(7))for a in range(11)]
print(result)
```
